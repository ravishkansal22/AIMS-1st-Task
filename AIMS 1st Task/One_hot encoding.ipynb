```python
import pandas as pd
import numpy as np

def one_hot_encode():
    data = input ("Enter categorical variables:").split(',')
    data = [x.strip() for x in data]
    categories=[]
    for item in data:
        if item not in categories:
            categories.append(item)
    print("Categories:", categories)
    print("One_hot_encoding: ")

    for value in data:
        row = []
        for cat in categories:
            if value == cat:
                row.append(1)
            else:
                row.append(0)
        print(f"{value}: {row}")
```

```python
one_hot_encode()
```
