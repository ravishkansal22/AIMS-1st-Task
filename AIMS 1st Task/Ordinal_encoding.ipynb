```python
import pandas as pd 
import numpy as np

def ordinal_encode():
    data = input ("Enter training categorical variable: ").split(',')
    data = [x.lower() for x in data]
    unique_values = []
    encoding_dict = {}
    counter = 1
    for item in data:
        if item not in unique_values:
            unique_values.append(item)
            encoding_dict[item] = counter
            counter += 1
    encoded_data = [encoding_dict[item] for item in data]
    print("Original Data: ", data)
    print("Encoding_dict: ", encoding_dict)
    print("Encoded Data: ", encoded_data)
```

```python
ordinal_encode()
```
