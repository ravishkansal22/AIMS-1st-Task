```python
import pandas as pd
import numpy as np

data = {
    'age': [22, np.nan, 28, 28, np.nan, 31],
    'salary': [55000, 65000, np.nan, 53000, 49000, np.nan],
    'city': ['New Delhi', 'Mumbai', np.nan, 'Indore', 'Vadodara', np.nan]
}

df = pd.DataFrame(data)

for mode_imputation in ['age', 'salary']:
    mode_value = df[mode_imputation].mode()[0]
    df[mode_imputation] = df[mode_imputation].fillna(mode_value)

df['city'] = df['city'].fillna('Unknown')
print(df)
```
