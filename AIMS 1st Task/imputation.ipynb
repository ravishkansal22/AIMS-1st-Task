```python
import pandas as pd
import numpy as np
data = {
    'age': [22, np.nan, 28, 28, np.nan, 31],
    'salary': [55000, 65000, np.nan, 53000, 49000, np.nan],
    'city': ['New Delhi', 'Mumbai', np.nan, 'Indore', 'Vadodara', np.nan]
}

df = pd.DataFrame(data)

def imputation():
    dict = {
        "Mean Imputation": "1",
        "Median Imputation": "2",
        "Mode Imputation": "3",
    }
    print("For mean imputation, enter 1\nFor median imputation, enter 2\nFor mode imputation, enter 3")
    a = int(input("Which imputation method would you like to use? \n"))
    if (a == 1):

        for mean_imputation in ['age', 'salary']:
            mean_value = df[mean_imputation].mean()
            df[mean_imputation] = df[mean_imputation].fillna(mean_value)

        df['city'] = df['city'].fillna('Unknown')
        print(df)
    elif (a == 2):

        for median_imputation in ['age', 'salary']:
            median_value = df[median_imputation].median()
            df[median_imputation] = df[median_imputation].fillna(median_value)

        df['city'] = df['city'].fillna('Unknown')
        print(df)
    elif (a == 3):

        for mode_imputation in ['age', 'salary']:
            mode_value = df[mode_imputation].mode()
            df[mode_imputation] = df[mode_imputation].fillna(mode_value)

        df['city'] = df['city'].fillna('Unknown')
        print(df)
    else :
        print("Invalid input. Please select a valid imputation method.")
```

```python
imputation()
```
